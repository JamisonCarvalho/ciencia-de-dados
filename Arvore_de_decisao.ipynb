# Arvóre de decisão

```python
from sklearn import tree

dados = [[0,0], [1,1],
         [2,0], [1,3],
         [2,6], [3,2]]
rotulos = [0,1,
           0,1,
           1,1]
```

### Treinando o modelo

```python
classificador = tree.DecisionTreeClassifier()
classificador = classificador.fit(dados, rotulos)
```

### Predição com o modelo

```python
teste = [1., 2.]
resultado = classificador.predict([teste])
print(f'O dado {teste} foi classificado como {resultado}')
```

### Exibir Árvore de classificação

```python
tree.plot_tree(classificador
               )
```
