# Análise discriminante

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

### Dados e grupos

```python
dados = np.array([[-1.5, -1.5], [-2.3, -2.1],
                  [-3.1, -2.7], [1.6, 1.2],
                  [2.3, 1.5], [3.1, 2.5]])

grupos = np.array([1,1,
                   1,2,
                   2,2])
```

## Método de análise linear disciminante

```python
clf = LinearDiscriminantAnalysis()
clf.fit(dados, grupos)
```

### Testar o método

```python
dados_teste = [0.0, 0.0]
grupo_resultado = clf.predict([dados_teste])
print(f'Os dados de teste {dados_teste} pertencem ao grupo {grupo_resultado}')
```

### Imprimir os dados no gráfico

```python
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(
    dados[:,0], dados[:,1],
    c= grupos,
    cmap = 'rainbow',
    alpha = 0.9,
    edgecolor = 'b')
plt.scatter( dados_teste[0], dados_teste[1],
    c= 'yellow',
    linewidths = 2,
    marker = "^",
    edgecolor = "red",
    s = 200)
plt.title("Análise Linear Discriminante")
plt.show()
```
